## 1.1 Problem Statement

Manufacturing defect detection is a critical task in industrial quality control. Traditional manual visual inspection:
- Is slow and expensive
- Is inconsistent across inspectors
- Cannot scale to high-speed production lines

**Goal:** Build a deep learning pipeline that:
1. Classifies product images as **defective** or **non-defective**
2. **Localizes** defect regions using Grad-CAM
3. Provides **anomaly scores** using autoencoders
4. Deploys as a real-time inspection web app

## 1.2 Research Questions
- Which CNN architecture achieves best precision-recall tradeoff for defect detection?
- Can autoencoders detect novel defect types unseen during training?
- How accurately does Grad-CAM localize defect regions compared to ground truth masks?
- What is the acceptable latency for real-time IoT integration?


## 1.3 Literature Review

| Paper | Method | Key Finding |
|-------|--------|-------------|
| Roth et al. (2022) — PatchCore | Memory bank of patch features | SOTA on MVTec, 99.1% AUROC |
| Li et al. (2021) — CutPaste | Self-supervised data augmentation | 95.2% AUROC without anomaly samples |
| Zavrtanik et al. (2021) — DRAEM | Reconstructive subspace learning | 98.0% AUROC |
| Selvaraju et al. (2017) — Grad-CAM | Gradient-weighted class activation maps | Visual explanations for CNN decisions |
| Akcay et al. (2019) — GANomaly | GAN-based anomaly detection | Strong on texture defects |

**Our approach:** Combine supervised CNN classification (ResNet-50) + unsupervised autoencoder anomaly scoring + Grad-CAM explainability — a practical ensemble for industrial deployment.

## 1.4 Evaluation Metrics

Given class imbalance (more normal than defective samples), we use:
- **AUROC** — primary metric, threshold-independent
- **F1 Score** — harmonic mean of precision and recall
- **Precision & Recall** — tradeoff analysis (missing a defect = costly)
- **IoU** — for Grad-CAM vs ground truth mask overlap

In manufacturing, **recall is more important than precision** — it's worse to miss a defect than to flag a good part.
